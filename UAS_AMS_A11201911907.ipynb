### **Import library yang akan digunakan**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

### **Import dataset**

```python
df = pd.read_csv("emails.csv")
df.head(5)
df.isnull().sum()
df.describe()
df.corr()
```

## **Membuat Model NB**

Dalam proyek ini, kami mengklasifikasikan surat yang diketik oleh pengguna sebagai 'Spam' atau 'Bukan Spam'. Dataset asli kami adalah folder dengan 5172 file teks yang berisi email.

Jika sekarang kita mendapatkan email tes, dan itu berisi ketiga kata di atas, ada kemungkinan tinggi bahwa itu adalah email 'Spam'.

Algoritma yang paling efektif untuk masalah pengklasifikasian teks adalah algoritma Naive Bayes, yang bekerja pada teorema Bayes klasik. Teorema ini bekerja pada setiap kata individu dalam data tes untuk membuat prediksi (probabilitas kondisional dengan probabilitas yang lebih tinggi adalah hasil yang diprediksi).

Katakanlah email tes kami (S) adalah, "Anda telah memenangkan sebuah lottery"

BAGAMANA NAIVE BAYES BEKERJA PADA DATA INI -->

P(S) = P('You') P('have') P('won') P('a') P('lottery') __ 1

Oleh Karena itu, P(S|Spam) = P('You'|Spam) P('have'|Spam) P('won'|Spam) P('a'|Spam) P('lottery'|Spam) __ 2

Perhitungan yang sama untuk P (S | Bukan_Spam) __ 3

Jika 2> 3, maka 'Spam' Jika tidak, 'Bukan_Spam'.

APA JIKA PROBALITASNYA NOL? Inilah konsep Lapisan Penyamaran, di mana P (kata-kata) = (jumlah kata + 1) / (jumlah total kata + jumlah kata unik)

Di sini, eksperimen dilakukan pada pengklasifikasi Naive Bayes Multinomial yang ada (di bawah scikit-learn). Untuk memahami lebih lanjut bagaimana Naive Bayes bekerja dengan baik untuk pengklasifikasian teks, akan digunakan klasifikator standar lainnya, SVC, untuk melihat bagaimana kinerja kedua model.

APAKAH MODEL ENSEMBLE AKAN BEKERJA LEBIH BAIK? Marilah kita lihat. Kami akan menggunakan Random Forests untuk membandingkan.

```python
X = df.iloc[:,1:3001]
X
```

```python
Y = df.iloc[:,-1].values
Y
```

### **Membuat variabel train dan test**

```python
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size = 0.25)
```

### **Perhitungan Naive Bayes**

```python
mnb = MultinomialNB(alpha=1.9)
mnb.fit(train_x,train_y)
y_pred1 = mnb.predict(test_x)
print("Accuracy Score for Naive Bayes : ", accuracy_score(y_pred1,test_y))
```

## **Support Vector Machines**

Support Vector Machines adalah algoritma yang paling dicari untuk masalah klasifikasi klasik. SVMs bekerja pada algoritma Maximal Margin, yaitu, mencari margin atau ambang batas maksimal antara vektor dukungan dari dua kelas (dalam klasifikasi biner). Mesin Vektor Pendukung yang paling efektif adalah kelasifikator margin maksimal lunak, yang memungkinkan satu kesalahan klasifikasi, yaitu, model dimulai dengan bias rendah (performa sedikit buruk) untuk memastikan variansi rendah kemudian.

```python
svc = SVC(C=1.0,kernel='rbf',gamma='auto')
svc.fit(train_x,train_y)
y_pred2 = svc.predict(test_x)
print("Accuracy Score for SVC : ", accuracy_score(y_pred2,test_y))
```

*Seperti yang diharapkan, performa SVM sedikit lebih buruk daripada Multinomial Naive Bayes.*

## **Random Forests (Bagging)**
Metode ensemble mengubah model yang lemah menjadi yang sangat kuat. Mari kita lihat apakah model ensemble dapat melakukan lebih baik daripada Naive Bayes.

```python
rfc = RandomForestClassifier(n_estimators=100,criterion='gini')
rfc.fit(train_x,train_y)
y_pred3 = rfc.predict(test_x)
print("Accuracy Score of Random Forest Classifier : ", accuracy_score(y_pred3,test_y))
```

*Seperti yang diharapkan, Random Forest Classifier melakukan yang terbaik di antara ketiganya. **Decision tree classifiers** adalah kelasifikator yang sangat baik. **Random forest** adalah model ensemble populer yang menggunakan **forest of decision trees**. Jadi, jelas, menggabungkan akurasi dari 100 trees (seperti n_estimators=100 di sini), akan membuat model yang kuat.*
